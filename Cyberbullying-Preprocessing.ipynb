```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# Function to preprocess the text
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Remove hashtags
    text = re.sub(r'#\w+', '', text)

    # Remove mentions
    text = re.sub(r'@\w+', '', text)

    # Remove emojis
    text = re.sub(r'[^\w\s,]', '', text)

    # Remove special characters (except apostrophes) and numbers
    text = re.sub(r'[^a-zA-Z\s\']', '', text)

    # Replace "rt" with "retweet"
    text = re.sub(r'\brt\b', 'retweet', text)

    text = re.sub(r'\bbullies\b', 'bully', text)
    text = re.sub(r'\bbullied\b', 'bully', text)
    text = re.sub(r'\bbullying\b', 'bully', text)

    text = re.sub(r'\bmuslims\b', 'muslim', text)

    text = re.sub(r'\bgirls\b', 'girl', text)

    text = re.sub(r'\bjokes\b', 'joke', text)

    text = re.sub(r'\bu\b', 'you', text)
    text = re.sub(r'\bfucking\b', 'fuck', text)
    text = re.sub(r'\bidiots\b', 'idiot', text)
    text = re.sub(r'\bniggers\b', 'nigger', text)


    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    return filtered_tokens

# Read the data from the CSV file
df = pd.read_csv("Cyberbullying.csv")

# Apply the preprocessing function to each tweet
df['processed_tweet'] = df['tweet_text'].apply(preprocess_text)


output_file_path = 'Cyberbully_preprocessed.csv'
df.to_csv(output_file_path, index=False)
```
